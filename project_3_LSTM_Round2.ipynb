```python
# Initial imports
import pandas as pd
import seaborn as sns


from pathlib import Path
```

```python
# 1. Weather DataFrame from NYC. 
weather_df = pd.read_csv(Path('data/weather_nyc_final.csv'), parse_dates=True, index_col='date')
weather_df.sort_index(ascending=True, inplace=True)

#weather_df.head()
```

```python
# 2. Crypto DataFrame
crypto_df = pd.read_csv(Path('crypto.csv'), parse_dates=True, index_col='Date')
crypto_df.sort_index(ascending=True, inplace=True)

#crypto_df.head()
```

```python
# 3. Energy Stock Ticker
stock_energy_df = pd.read_csv(Path('stocks_final.csv'), parse_dates=True, index_col='Date')
stock_energy_df.sort_index(ascending=True, inplace=True)
stock_energy_df.head()
```

```python
# 4. Commodities
commodities_df = pd.read_csv(Path('commodities.csv'), parse_dates=True, index_col='Date')
commodities_df.sort_index(ascending=True, inplace=True)
commodities_df.head()
```

```python
# Define all the different data frames into a list
combined_frames = [weather_df,
              crypto_df,
              stock_energy_df,
              commodities_df]

# Concatenate all the dataframes into one
combined_df = pd.concat(combined_frames, axis=1, join='inner', ignore_index=False)
combined_df.head()
```

```python
# Copy combined df to create ALL potential features dataset
potential_feature_df = combined_df.copy()
potential_feature_df.drop(
                                    ['BTC Volume',
                                    'BTC New Coins Mined',
                                    'BTC Active Addresses',
                                    'BTC Transactions',
                                    'BTC Daily Transactions Change (%)',
                                    'icln_Open',
                                    'icln_High',
                                    'icln_Low',
                                    'xlu_Open',
                                    'xlu_High',
                                    'xlu_Low',
                                    'vpu_Open',
                                    'vpu_High',
                                    'vpu_Low',
                                    'ryu_Open',
                                    'ryu_High',
                                    'ryu_Low',
                                    'futy_Open',
                                    'futy_High',
                                    'futy_Low',
                                    'Corn Fut Open',
                                    'Corn Fut High',
                                    'Corn Fut Low',
                                    'SB Fut Open',
                                    'SB Fut High',
                                    'SB Fut Low',
                                   ],
                                axis=1, inplace=True)

potential_feature_df.head()
```

```python
potential_feature_df.info()
```

```python
# Copy combined df to create ALL potential features dataset
ml_df = potential_feature_df.copy()
ml_df.drop(
                                    ['BTC Daily Volume Change (%)',
                                    'BTC New Coins Mined (%)',
                                    'BTC Daily Active Addresses Change (%)',
                                    'icln_Close',
                                    'icln_Volume',
                                    'xlu_Close',
                                    'xlu_Volume',
                                    'vpu_Close',
                                    'vpu_Volume',
                                    'ryu_Close',
                                    'ryu_Volume',
                                    'futy_Close',
                                    'futy_Volume',
                                    'Nat Gas Open',
                                    'Nat Gas Close',
                                    'Nat Gas Volume',
                                    'Crude Oil Open',
                                    'Crude Oil Close',
                                    'Crude Oil Volume',
                                    'Corn Fut Close',
                                    'Corn Fut Adj Close',
                                    'Corn Fut Volume',
                                    'Corn Fut pctChange',
                                    'SB Fut Close',
                                    'SB Fut Adj Close',
                                    'SB Fut Volume',
                                    'SB Fut pctChange'
                                   ],
                                axis=1, inplace=True)

ml_df.head()
```

```python
# Check for no null values
ml_df.info()
```

```python
# Adding predict column looking out 1 day
ml_df['pred'] = ml_df.icln_Adj_Close.shift(-1)
ml_df = ml_df.dropna()
ml_df.head()
```

```python
# Grab data values
values = ml_df.values

# Check to make sure all data is float
values = values.astype('float64')
values
```

```python
# Import libraries and scale data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))

scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)
scaled = pd.DataFrame(scaled)
scaled.head()
```

```python
# Create the lookback formula and the arrays
def ts (ml_df, look_back = 60, pred_col = 1):
    t = ml_df.copy()
    t["id"] = range(1, len(t)+1)
    t = t.iloc[:-look_back, :]
    t.set_index('id', inplace=True)
    pred_value = ml_df.copy()
    pred_value = pred_value.iloc[look_back:, pred_col]
    pred_value.columns = ["Pred"]
    pred_value = pd.DataFrame(pred_value)
    
    pred_value["id"] = range (1, len(pred_value)+1)
    pred_value.set_index('id', inplace = True)
    final_df = pd.concat([t, pred_value], axis=1)
    
    return final_df
```

```python
arr_df = ts(scaled, 60,0)
arr_df.fillna(0, inplace=True)

arr_df.columns = ['v1(t-60)', 
                 'v2(t-60)',
                 'v3(t-60)',
                 'v4(t-60)',
                 'v5(t-60)',
                 'v6(t-60)',
                 'v7(t-60)',
                 'v8(t-60)',
                 'v9(t-60)',
                 'v10(t-60)',
                 'v11(t-60)',
                 'v12(t-60)',
                 'v13(t-60)',
                 'v14(t-60)',
                 'v15(t-60)',
                 'v16(t-60)',
                 'v17(t-60)',
                 'v18(t-60)',
                 'v19(t-60)',
                 'v20(t-60)',
                 'v1(t)']

arr_df.head()
```

```python
# Split data sets
values = arr_df.values
train_sample = int(len(ml_df) * 0.8)
train = values [: train_sample, :]
test = values [train_sample:, :]
```

```python
# Split into inputs and outputs
xtrain, ytrain = train[:, :-1], train[:, -1]
xtest, ytest = test[:, :-1], test[:, -1]

print(xtrain.shape, ytrain.shape, xtest.shape, ytest.shape)
```

```python
# Reshaping to make data set 3d
xtrain = xtrain.reshape((xtrain.shape[0], 1, xtrain.shape[1]))
xtest = xtest.reshape((xtest.shape[0], 1, xtest.shape[1]))
print(xtrain.shape, ytrain.shape, xtest.shape, ytest.shape)
```

```python
import tensorflow as tf
```

```python
# Build the model
model_lstm = tf.keras.Sequential()
model_lstm.add(tf.keras.layers.LSTM(50, return_sequences=True, input_shape = (xtrain.shape[1], xtrain.shape[2])))
model_lstm.add(tf.keras.layers.LSTM(units=30, return_sequences=True))
model_lstm.add(tf.keras.layers.LSTM(units=30))
model_lstm.add(tf.keras.layers.Dense(units=1))
model_lstm.compile(loss = 'mae', optimizer = 'adam')
model_lstm.summary()
history_lstm = model_lstm.fit(xtrain, ytrain, epochs = 30,        batch_size=32, validation_data = (xtest, ytest), shuffle=False)
```

```python
# Import plots from matplotlib
import matplotlib.pyplot as plt

# Plot the training
plt.plot(history_lstm.history['loss'], label='train')
plt.plot(history_lstm.history['val_loss'], label='test')
plt.legend()
plt.show()
```

```python
# make prediction
yhat = model_lstm.predict(xtest)
xtest = xtest.reshape((xtest.shape[0], xtest.shape[2]))
```

```python
from numpy import concatenate
# invert scaling for forecast
inv_yhat = concatenate((yhat, xtest[:, 1:]), axis=1)
inv_yhat = scaler.inverse_transform(inv_yhat)
inv_yhat = inv_yhat[:,0]
```

```python
# invert scaling for actual
ytest = ytest.reshape((len(ytest), 1))
inv_y = concatenate((ytest, xtest[:, 1:]), axis=1)
inv_y = scaler.inverse_transform(inv_y)
inv_y = inv_y[:,0]
```

```python
import math
import sklearn
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

```

```python
print("Mean absolute error (MAE): %f" % sklearn.metrics.mean_absolute_error(inv_y, inv_yhat))
print("Mean squared error (MSE): %f" % sklearn.metrics.mean_squared_error(inv_y, inv_yhat))
print("Root mean squared error (RMSE): %f" % math.sqrt(sklearn.metrics.mean_squared_error(inv_y, inv_yhat)))
print("R square (R^2): %f" % sklearn.metrics.r2_score(inv_y, inv_yhat))
```

```python
# Show actual vs. predicted
plt.plot(inv_y, label = 'Actual')
plt.plot(inv_yhat, label = 'Predicted')
plt.legend()
plt.show()
```
