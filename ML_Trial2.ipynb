```python
# Initial imports
import pandas as pd
import seaborn as sns
import numpy as np
from pathlib import Path
```

```python
# 1. Weather DataFrame from NYC. 
weather_df = pd.read_csv(Path('weather_nyc_final.csv'), parse_dates=True, index_col='date')
weather_df.sort_index(ascending=True, inplace=True)

#weather_df.head()
```

```python
# 2. Crypto DataFrame
crypto_df = pd.read_csv(Path('crypto.csv'), parse_dates=True, index_col='Date')
crypto_df.sort_index(ascending=True, inplace=True)

#crypto_df.head()
```

```python
# 3. Energy Stock Ticker
stock_energy_df = pd.read_csv(Path('stocks.csv'), parse_dates=True, index_col='Date')
stock_energy_df.sort_index(ascending=True, inplace=True)
stock_energy_df.head()
```

```python
# 4. Commodities
commodities_df = pd.read_csv(Path('commodities.csv'), parse_dates=True, index_col='Date')
commodities_df.sort_index(ascending=True, inplace=True)
commodities_df.head()
```

```python
# Define all the different data frames into a list
combined_frames = [weather_df,
              crypto_df,
              stock_energy_df,
              commodities_df]

# Concatenate all the dataframes into one
combined_df = pd.concat(combined_frames, axis=1, join='inner', ignore_index=False)
combined_df.head()
```

```python
combined_df.info()
```

```python
# Copy combined df to create feature dataset
ml_dataset_df = combined_df.copy()
ml_dataset_df.drop(
                                    ['BTC Volume',
                                    'BTC New Coins Mined',
                                    'BTC Active Addresses',
                                    'BTC Transactions',
                                    'BTC Daily Transactions Change (%)',
                                    'xle_Open',
                                    'xle_High',
                                    'xle_Low',
                                    'xle_Close',
                                    'icln_Open',
                                    'icln_High',
                                    'icln_Low',
                                    'xop_Open',
                                    'xop_High',
                                    'xop_Low',
                                    'xop_Close',
                                    'amlp_Open',
                                    'amlp_High',
                                    'amlp_Low',
                                    'amlp_Close',
                                    'vde_Open',
                                    'vde_Low',
                                    'vde_Close',
                                    'Nat Gas Open',
                                    'Nat Gas High',
                                    'Nat Gas Low',
                                    'Nat Gas Close',
                                    'Crude Oil Open',
                                    'Crude Oil High',
                                    'Crude Oil Close',
                                    'Corn Fut Open',
                                    'Corn Fut High',
                                    'Corn Fut Low',
                                    'Corn Fut Close',
                                    'SB Fut Open',
                                    'SB Fut High',
                                    'SB Fut Low',
                                    'SB Fut Close'
                                   ],
                                axis=1, inplace=True)

ml_dataset_df.head()
```

```python
# Create a dataset for just xle
xle_ml_df = ml_dataset_df.copy()
xle_ml_df.drop(['icln_Adj_Close',
                'icln_Volume',
                'icln_pctChange',
                'xop_Adj_Close',
                'xop_Volume',
                'xop_pctChange',
                'amlp_Adj_Close',
                'amlp_Volume',
                'amlp_pctChange',
                'vde_Adj_Close',
                'vde_Volume',
                'vde_pctChange'],
              axis=1, inplace=True)

xle_ml_df.head()
```

```python
xle_ml_df.dropna(axis=0, inplace=True)
xle_ml_df.head()
```

```python
# Random Forest classification to check for feature importance

# Initial Random Forest Imports
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from matplotlib import pyplot
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
%matplotlib inline
import pydotplus
from IPython.display import Image
```

```python
# Define Features for xle dataset
X = xle_ml_df.copy()
X.drop('xle_Adj_Close', axis=1, inplace=True)
X.head()
```

```python
len(X)
```

```python
y = xle_ml_df['xle_Adj_Close']
```

```python
# Splitting into Train and Test sets
split = int(0.7 * len(X))

X_train = X[: split]
X_test = X[split:]

y_train = y[: split]
y_test = y[split:]

# Creating StandardScaler instance
scaler = StandardScaler()

# Fitting Standard Scaller
X_scaler = scaler.fit(X_train)

# Scaling data
X_train = X_scaler.transform(X_train)
X_test = X_scaler.transform(X_test)
```

```python
len(X_test)
```

```python
len(y_test)
```

```python
#define model (try to n_features function for automation of df drop)
model = LinearRegression()
#fit model
model.fit(X_train, y_train)
```

```python
importance = model.coef_
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
```

```python
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

```python
importances = importance
```

```python
#sorting importances
sorted(zip(importances, X.columns), reverse=True)
```

```python
importances_df =pd.DataFrame(sorted(zip(importances, X.columns), reverse=True))
importances_df
```

```python
# Visualize the features by importance
importances_df.set_index(importances_df[1], inplace=True)
importances_df.drop(columns=1, inplace=True)
importances_df.rename(columns={0: 'Feature Importances'}, inplace=True)
importances_sorted = importances_df.sort_values(by='Feature Importances')
importances_sorted.plot(kind='barh', color='lightgreen', title= 'Features Importances', legend=False)
```

```python
# Making predictions using the testing data
predictions = model.predict(X_test)
predictions
```

```python
Results = y_test
Out_of_sample_results = pd.DataFrame({"Actual_Results": y_test,
                   "Predicted Returns": predictions})
```

```python
Out_of_sample_results.plot()
```

```python
#out of sample performance
out_of_sample_mse = mean_squared_error(
    Out_of_sample_results["Actual_Results"],
    Out_of_sample_results["Predicted Returns"]
)

# Using that mean-squared-error, calculate the root-mean-squared error (RMSE):

out_of_sample_rmse = np.sqrt(out_of_sample_mse)
r2 = r2_score(Out_of_sample_results["Actual_Results"], Out_of_sample_results["Predicted Returns"])

print(f"Out-of-sample Root Mean Squared Error (RMSE): {out_of_sample_rmse}")
print(f"Out-of-Sample r2: {r2}")
```

```python
#AutoRegressor library components
from sklearn.datasets import make_regression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeRegressor
from sklearn.pipeline import Pipeline
from numpy import mean
from numpy import std
```

```python
#Recursive Feature Elimination logic
rfe = RFE(estimator=DecisionTreeRegressor(), n_features_to_select=5)
pipeline = Pipeline(steps=[('s',rfe),('m',model)])
```

```python
#Evaluate Model
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(pipeline, X_train, y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1, error_score='raise')
```

```python
#RFE performance. MAE = Mean Absolute Error
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
#re-running model with recursive feature elimination prediction
pipeline.fit(X_train, y_train)
yhat = pipeline.predict(X_test)
```

```python
oos_pipeline_results = pd.DataFrame({"Actual_Results": y_test,
                   "Pipeline Predicted Returns": yhat})
```

```python
oos_pipeline_results.head()
```

```python
#out of sample PIPELINE  performance
oos_pipeline_mse = mean_squared_error(
    oos_pipeline_results["Actual_Results"],
    oos_pipeline_results["Pipeline Predicted Returns"]
)

# Using that mean-squared-error, calculate the root-mean-squared error (RMSE):

oos_pipeline_mse = np.sqrt(oos_pipeline_mse)
r2 = r2_score(oos_pipeline_results["Actual_Results"], oos_pipeline_results["Pipeline Predicted Returns"])

print(f"Out-of-sample Pipeline Root Mean Squared Error (RMSE): {oos_pipeline_mse}")
print(f"Out-of-Sample Pipeline r2: {r2}")
```

```python
oos_pipeline_results.plot()
```
