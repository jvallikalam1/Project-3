```python
# Initial imports
import pandas as pd
from pathlib import Path
```

```python
#https://www.wunderground.com/history/daily/us/ny/new-york-city/KLGA
# Due to the fact that we couldnt get historical weather data with an API or extract - 
# We had to take a manual approach and decided to collect the data by creating our own dataset

df = pd.read_csv(Path('data/weather_nyc.csv'), parse_dates=True, index_col='date')

#sort the data by date
df.sort_index(ascending=True, inplace=True)

df.head()
```

```python
# Creting a new column 'Month' - based on the actual Month of the date (index)
df['month'] = df.index.month_name()

# Calculating the average between Max Temp and Min Temp
df['temp_average'] = (df['temp_max'] + df['temp_min']) / 2

# Remove irrelvant columns
df_final = df.drop(['temp_max', 'temp_avg', 'temp_min', 'dew_point_max', 'dew_point_avg', 'dew_point_min', 'humidity_max', 'humidity_avg', 'humidity_min', 'wind_max', 'wind_avg', 'wind_min'], axis = 1)
```

```python
# https://www.currentresults.com/Weather/New-York/Places/new-york-city-temperatures-by-month-average.php

avg_temp_dict = {
    'January': 32.5,
    'February': 35.5, 
    'March': 42.5,
    'April': 53,
    'May': 62.5,
    'June': 71.5,
    'July': 76.5,
    'August': 75.5,
    'September': 68,
    'October': 57,
    'November': 48,
    'December': 37.5
}
```

```python
df_final['temp_average_decade'] = df_final['month'].map(avg_temp_dict)   
```

```python
df_final
```

```python
# Calculating the difference between daily average temp and decade average temp for the specific month
df_final['temp_diff'] = (df_final['temp_average'] - df_final['temp_average_decade'])
```

```python
# Remove irrelvant columns
df_final_temp = df_final.drop(['month', 'temp_average', 'temp_average_decade'], axis = 1)
```

```python
# save filtered Weather data for NYC

df_final_temp.to_csv('data/weather_nyc_final.csv', index=True)
```
