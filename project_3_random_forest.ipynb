```python
# Initial imports
import pandas as pd
import seaborn as sns

from pathlib import Path
```

```python
# 1. Weather DataFrame from NYC. 
weather_df = pd.read_csv(Path('weather_nyc_final.csv'), parse_dates=True, index_col='date')
weather_df.sort_index(ascending=True, inplace=True)

#weather_df.head()
```

```python
# 2. Crypto DataFrame
crypto_df = pd.read_csv(Path('crypto.csv'), parse_dates=True, index_col='Date')
crypto_df.sort_index(ascending=True, inplace=True)

#crypto_df.head()
```

```python
# 3. Energy Stock Ticker
stock_energy_df = pd.read_csv(Path('stocks.csv'), parse_dates=True, index_col='Date')
stock_energy_df.sort_index(ascending=True, inplace=True)
stock_energy_df.head()
```

```python
# 4. Commodities
commodities_df = pd.read_csv(Path('commodities.csv'), parse_dates=True, index_col='Date')
commodities_df.sort_index(ascending=True, inplace=True)
commodities_df.head()
```

```python
# Define all the different data frames into a list
combined_frames = [weather_df,
              crypto_df,
              stock_energy_df,
              commodities_df]

# Concatenate all the dataframes into one
combined_df = pd.concat(combined_frames, axis=1, join='inner', ignore_index=False)
combined_df.head()
```

```python
combined_df.info()
```

```python
# Copy combined df to create feature dataset
ml_dataset_df = combined_df.copy()
ml_dataset_df.drop(
                                    ['BTC Volume',
                                    'BTC New Coins Mined',
                                    'BTC Active Addresses',
                                    'BTC Transactions',
                                    'BTC Daily Transactions Change (%)',
                                    'xle_Open',
                                    'xle_High',
                                    'xle_Low',
                                    'xle_Close',
                                    'icln_Open',
                                    'icln_High',
                                    'icln_Low',
                                    'xop_Open',
                                    'xop_High',
                                    'xop_Low',
                                    'xop_Close',
                                    'amlp_Open',
                                    'amlp_High',
                                    'amlp_Low',
                                    'amlp_Close',
                                    'vde_Open',
                                    'vde_Low',
                                    'vde_Close',
                                    'Nat Gas Open',
                                    'Nat Gas High',
                                    'Nat Gas Low',
                                    'Nat Gas Close',
                                    'Crude Oil Open',
                                    'Crude Oil High',
                                    'Crude Oil Close',
                                    'Corn Fut Open',
                                    'Corn Fut High',
                                    'Corn Fut Low',
                                    'Corn Fut Close',
                                    'SB Fut Open',
                                    'SB Fut High',
                                    'SB Fut Low',
                                    'SB Fut Close'
                                   ],
                                axis=1, inplace=True)

ml_dataset_df.head()
```

```python
# Create a dataset for just xle
xle_ml_df = ml_dataset_df.copy()
xle_ml_df.drop(['icln_Adj_Close',
                'icln_Volume',
                'icln_pctChange',
                'xop_Adj_Close',
                'xop_Volume',
                'xop_pctChange',
                'amlp_Adj_Close',
                'amlp_Volume',
                'amlp_pctChange',
                'vde_Adj_Close',
                'vde_Volume',
                'vde_pctChange'],
              axis=1, inplace=True)

xle_ml_df.head()
```

```python
xle_ml_df.dropna(axis=0, inplace=True)
xle_ml_df.head()
```

```python
# Random Forest classification to check for feature importance

# Initial Random Forest Imports
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
%matplotlib inline
import pydotplus
from IPython.display import Image
```

```python
# Define Features for xle dataset
X = xle_ml_df.copy()
X.drop('xle_pctChange', axis=1, inplace=True)
X.head()
```

```python
y = xle_ml_df['xle_pctChange'].ravel()
y[:5]
```

```python
# Splitting into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

# Creating StandardScaler instance
scaler = StandardScaler()

# Fitting Standard Scaller
X_scaler = scaler.fit(X_train)

# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

```

```python
# Create a random forest classifier
rf_model = RandomForestClassifier(n_estimators=500, random_state=78)

# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data
predictions = rf_model.predict(X_test_scaled)
```
